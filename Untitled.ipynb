```python
def specToXYZ(betha, Lw, x_, y_, z_):
    XYZ = [0,0,0]
    XYZ[0] = np.sum(x_ * betha * Lw)
    XYZ[1] = np.sum(y_ * betha * Lw)
    XYZ[2] = np.sum(z_ * betha * Lw)
        
    return XYZ

#converte xyz para srgb
def XYZ_sRGB(XYZ):
    
    def gamma_sRGB(x):
        if x < 0.0031308:
            return 12.92 * x
        else:
            return 1.055*pow(x, 1/2.4) - 0.055
        
    r,g,b = 0,0,0
    
    r = 3.2404542*XYZ[0] - 1.5371385*XYZ[1] - 0.4985314*XYZ[2]
    g = -0.9692660*XYZ[0] + 1.8760108*XYZ[1] + 0.0415560*XYZ[2]
    b =  0.0556434*XYZ[0] - 0.2040259*XYZ[1] + 1.0572252*XYZ[2]
        
    r = gamma_sRGB(r) * 255
    g = gamma_sRGB(g) * 255
    b = gamma_sRGB(b) * 255
    
    return [r,g,b]

#converte XYZ pra LAB usando D65
#CIE 2
def XYZ_LAB(XYZ):
    
    LAB = [0,0,0]
    
    var_X = XYZ[0] / 0.95047
    var_Y = XYZ[1] / 1.0
    var_Z = XYZ[2] / 1.08883
    
    if var_X > 0.008856:
        var_X = pow(var_X,(1/3))
    else:
        var_X = ( 7.787 * var_X ) + ( 16 / 116 )
        
    if var_Y > 0.008856:
        var_Y = pow(var_Y, (1/3))
    else:
        var_Y = ( 7.787 * var_Y ) + ( 16 / 116 )
        
    if var_Z > 0.008856:
        var_Z = pow(var_Z, (1/3))
    else:
        var_Z = ( 7.787 * var_Z ) + ( 16 / 116 )
        
    LAB[0] = ( 116 * var_Y ) - 16
    LAB[1] = 500 * ( var_X - var_Y )
    LAB[2] = 200 * ( var_Y - var_Z )
    
    return LAB
    
    
```

```python
def gera_beta(delta):
    betha = np.zeros((401))
    for i in range(delta):
        betha[i] = 1.0
    for i in range(delta, 401):
        betha[i] = 0.0
    return betha


def computeGamutPoints():
    
    df = pd.read_excel("all_1nm_data.xls", header=3)
    
    Lw = df['CIE D65']
    x_ = df["x bar"]
    y_ = df["y bar"]
    z_ = df["z bar"]
    
    Lw = Lw[80:481]
    x_ = x_[80:481]
    y_ = y_[80:481]
    z_ = z_[80:481]
        
    x_ = np.asarray(x_)
    y_ = np.asarray(y_)
    z_ = np.asarray(z_)
    Lw = np.asarray(Lw)
    
    k = 1.0/np.sum(y_ * Lw)
    Lw = Lw * k
    
    XYZPoints = [[0,0,0]]
    LABPoints = [[0,0,0]]
    sRGBPoints = [[0,0,0]]
    
    for delta in range(5,401,5):
        betha = gera_beta(delta)
        for i in range(0,401,1):
            betha = np.roll(betha,1)
            XYZ = specToXYZ(betha, Lw, x_, y_, z_)
            sRGB = XYZ_sRGB(XYZ)
            LAB = XYZ_LAB(XYZ)
            XYZPoints.append(XYZ)
            sRGBPoints.append(sRGB)
            LABPoints.append(LAB)

    return np.array(XYZPoints), np.array(sRGBPoints), np.array(LABPoints)
            
def draw(X,Y,Z, rgb,Xlabel, Ylabel, Zlabel):
    fig = plt.figure(figsize=(15,10))
    ax = fig.add_subplot(111,projection='3d')
    ax.scatter(X,Y,Z, color=rgb)
    ax.set_xlabel(Xlabel)
    ax.set_ylabel(Ylabel)
    ax.set_zlabel(Zlabel)
    plt.show()

#Limita valores entre 0 e 1
def clamp(x, xmin=0.0, xmax=1.0):
    if x < xmin:
        return xmin
    elif x > xmax:
        return xmax
    else:
        return x
```

```python
XYZArray, sRGBArray, LABA = computeGamutPoints()

rgb = []

for i in range(len(sRGBArray)):
    red = clamp(sRGBArray[i][0])
    green = clamp(sRGBArray[i][1])
    blue = clamp(sRGBArray[i][2])
    rgb.append([red,green,blue])
rgb = np.array(rgb)

print(XYZArray[:,2])
```

```python
print("Espaco XYZ")
draw(XYZArray[:,2], XYZArray[:,0],XYZArray[:,1], rgb, "CIE-z","CIE-X", "CIE-Y")
```
